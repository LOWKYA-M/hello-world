```
import pandas
import matplotlib.pyplot as plt
import numpy as np
fig= plt.figure(figsize=(20,10))
df=pandas.read_csv('/content/drive/My Drive/bitcoin.csv')
df['grossprofit']=df['Open']-df['Close']
plt.xlabel('date')
plt.ylabel('grossprofit')
d_values=np.array(df['Date'])
g_values=np.array(df['grossprofit'])
plt.plot(d_values,g_values,'go-')
plt.tick_params(axis="x", labelsize=20, labelrotation=90, labelcolor="red")
plt.tick_params(axis="y", labelsize=20, labelrotation=20, labelcolor="red")
plt.show()
df
```
